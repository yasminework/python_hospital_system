```python
###################################Pateint###############################
class Patient():
  def __init__(self,name,status):
    self.name=name
    self.status=status

############################specialization#########################################
class Spec():
  def __init__(self,spec_num):
    self.spec_num=spec_num
    self.patients=list()
    self.count=0

  def Add_patient(self,patient):
    if patient.status ==0:
      i=0
      while i<self.count:
        if self.patients[i].status==0:
          i+=1
          continue
        else:
          break
      self.patients.insert(i,patient)
    elif patient.status ==1:
      i=0
      while i<self.count:
        if self.patients[i].status!=2:
          i+=1
          continue
        else:
          break
      self.patients.insert(i,patient)  
    elif patient.status==2:
      self.patients.append(patient)
    self.count+=1


  def Print_All_Paient(self):
    for obj in self.patients:
      print("Patient:{} is {} ".format(obj.name,obj.status))


  def Next_Patient(self):
    if self.count==0:
      print("NO Patient found in this Specialization")
    else:
      print("patient {} please go to the DR".format(self.patients[0].name))
      self.count-=1
      self.patients.pop(0)

  def Remove_Patient(self):
    present=False
    name_inp=input("Please Enter Patient Name:")
    if self.count==0:
      print("NO Patient found in this Specialization")
    else:
      self.count-=1 
      for i, o in enumerate(self.patients):
        if o.name == name_inp:
          print("patient {} leaving".format(name_inp))
          del self.patients[i]
          self.Print_All_Paient()
          present=True
          break
      if present==False:
        print("No patient with name {} in this specialization ".format(name_inp))
 
##########################functions###################################
def search_spec(spec_num,spec_list):
  present = False
  spec=Spec(spec_num)
  for obj  in spec_list:
    if obj.count<10:
      if obj.spec_num!=spec_num:
        continue
      else:
        name=input("Enter Patient name: ")
        status=int(input("Enter Status: "))
        if status >=0 and status <=2:
          patient1=Patient(name,status)
          obj.Add_patient(patient1)
          print("adding patient",patient1.name,obj.count)
          present=True
          break
        else:
          print("Enter valid status")
    else:
      print("Sorry We cannot accept more patients in this specializations in this moment!")
      present=True
  if present== False:
    spec_list.append(spec)
    name=input("Enter Patient name: ")
    status=int(input("Enter Status: "))
    if status >=0 and status <=2:
      patient1=Patient(name,status)
      spec.Add_patient(patient1)
    else:
       print("Enter valid status")

def Get_patient(sp_inp,spec_list):
  present = False
  for obj  in spec_list:
    if obj.spec_num==sp_inp:
      obj.Next_Patient()
      present=True
  if present==False:
    print("No Specialization found")
def leaving_patient(sp_inp,spec_list):
  present = False
  for obj  in spec_list:
    if obj.spec_num==sp_inp:
      obj.Remove_Patient()
      present=True
  if present==False:
    print("No Specialization found")
#####################################################dummy######################################
def dumpy_function(p_list):
  s=Spec(1)
  for i in range(3):
    for j in range(3):
      p=Patient("patient"+str(j+i),j)
      s.Add_patient(p)
  s.Add_patient(Patient("patient10",1))
  p_list.append(s)
  s=Spec(2)
     
  for i in range(5):
    p=Patient("patient"+str(i),1)
    s.Add_patient(p)
  p_list.append(s)
 
  s=Spec(3)
  for i in range(3):
    p=Patient("patient"+str(i+1),1)
    s.Add_patient(p)
    p=Patient("patient"+str(2*i+1),2)
    s.Add_patient(p)
  p_list.append(s)



##########################################main###############################################      
spec_list=[]
dumpy_function(spec_list)  
while True:
  
  print("Program Options:\n 1)Add new patient \n 2)Print All Patient \n 3)Get Next Patient \n 4)Remove a leaving Patient \n 5)End the program ")
  inp=int(input("Enter your choise from 1-5: "))
  if inp==5:
    break
  if inp==1:
    spec_num=int(input("Enter Specialization: "))
    if  spec_num>=1 and spec_num<=20:
      search_spec(spec_num,spec_list)
    else:
      print("Enter a valid Specialization")
  if inp==2:
    for obj in spec_list:
     print("specialization:{} There are {} Patients.".format(obj.spec_num,obj.count))
     obj.Print_All_Paient()
  if inp==3:
    sp_inp=int(input("Enter Spetialization:"))
    if sp_inp>=1 and sp_inp<=20:
      Get_patient(sp_inp,spec_list)
  if inp==4:
    sp_inp=int(input("Enter Spetialization:"))
    if sp_inp>=1 and sp_inp<=20:
      leaving_patient(sp_inp,spec_list)
```
